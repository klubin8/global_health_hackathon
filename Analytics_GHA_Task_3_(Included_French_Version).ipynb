### Program costs, funding sources, and funding gaps disaggregated by activity, sub-activity and donor


* Activity related analysis

```python
## Activity related analysis
activity_Data = data[['Name of Activity.2','Amount of Finance Recevied','Gap']]

activity_Data = activity_Data.groupby(['Name of Activity.2'])['Amount of Finance Recevied','Gap'].sum().reset_index()

activity_Data['Total Program Costs (Activity Costs)'] = activity_Data[['Amount of Finance Recevied','Gap']].sum(axis=1)

activity_Datareport = activity_Data[['Name of Activity.2','Total Program Costs (Activity Costs)','Amount of Finance Recevied','Gap']]

activity_Datareport.rename(columns={'Name of Activity.2': 'Name of the Activity','Amount of Finance Recevied':'Amount of Funding Received'}, inplace=True)

activity_Datareport['Funding Received in Percentage'] = activity_Datareport['Amount of Funding Received']*100/activity_Datareport['Total Program Costs (Activity Costs)']

activity_Datareport
```

- Sub-activity related analysis

```python
#Select columns needed for sub-activity related analysis
subactivity_Data = data[['Name of Subactivity.1','Cost of Subactivity in GNF.1','Amount of Finance Recevied','Gap']]

#Select columns needed for sub-activity related analysis
subactivity_Data = data[['Name of Subactivity.1','Amount of Finance Recevied','Gap']]

subactivity_Data = subactivity_Data.groupby(['Name of Subactivity.1'])['Amount of Finance Recevied','Gap'].sum().reset_index()

subactivity_Data['Total Program Costs (Sub-activity Costs)'] = subactivity_Data[['Amount of Finance Recevied','Gap']].sum(axis=1)

subactivity_Datareport = subactivity_Data[['Name of Subactivity.1','Total Program Costs (Sub-activity Costs)','Amount of Finance Recevied','Gap']]

subactivity_Datareport.rename(columns={'Name of Subactivity.1': 'Name of the Sub-activity','Amount of Finance Recevied':'Amount of Funding Received'}, inplace=True)

subactivity_Datareport['Funding Received in Percentage'] = subactivity_Datareport['Amount of Funding Received']*100/subactivity_Datareport['Total Program Costs (Sub-activity Costs)']

subactivity_Datareport
```

- Donor Data Analysis

```python
donor_Data = data[[  'Cost of Subactivity in GNF.1', 'National Government', 'USAID', 'Sightsavers',
                     'Fondation Raoul Follereau France', 'Speak Up Africa', 'Ordre de Malte', 'WHO',
                     'Catholic Relief Services ','Institut de Recherche  et de Développement France ',
                     'Institut Pasteur  de  Guinée']]

donor_Data = donor_Data.sum().to_frame(name = None)#.rest_index()

donor_Data = donor_Data.rename(columns={0: 'Funding received in GNF'})

donor_Data[['Total cost of all programs in GNF']] = donor_Data[['Funding received in GNF']].iloc[0]

donor_Data['Funding gap in GNF'] = donor_Data['Total cost of all programs in GNF'] - donor_Data['Funding received in GNF']

donor_Data = donor_Data.iloc[1:,:]

#Re-arrange columns
donor_Data = donor_Data[['Total cost of all programs in GNF','Funding received in GNF','Funding gap in GNF']]

donor_Data = donor_Data.reset_index()

donor_Data = donor_Data.rename(columns={'index': 'Donor'})

donor_Data['Funding Received in Percentage'] = donor_Data['Funding received in GNF']*100/donor_Data['Total cost of all programs in GNF']

donor_Data
```

### Conversion from English to French language

```python
# activity_Datareport in French
model = EasyNMT("opus-mt")

activity_Datareport

# Create french version
activity_Datareport_French = activity_Datareport.copy()
activity_Datareport_French['Name of the Activity'] = activity_Datareport_French.apply(lambda row: model.translate(row["Name of the Activity"], target_lang="fr"), axis=1)
#activity_Datareport_French['Districts'] = activity_Datareport_French.apply(lambda row: model.translate(row["Districts"], target_lang="fr"), axis=1)

## Column rename
activity_Datareport_French = activity_Datareport_French.rename(columns={ 
                                                'Name of the Activity': model.translate('Name of the Activity', target_lang="fr"),
                                                'Total Program Costs (Activity Costs)' : model.translate('Total Program Costs (Activity Costs)', target_lang="fr"),
                                                'Amount of Funding Received' : model.translate('Amount of Funding Received', target_lang="fr") ,
                                                'Gap' : model.translate('Gap', target_lang="fr"),
                                                'Funding Received in Percentage' : model.translate('Funding Received in Percentage', target_lang="fr")})
activity_Datareport_French
```

```python
# Sub-activity Datareport in French
subactivity_Datareport 

# Create french version
subactivity_Datareport_French = subactivity_Datareport.copy()
subactivity_Datareport_French['Name of the Sub-activity'] = subactivity_Datareport_French.apply(lambda row: model.translate(row["Name of the Sub-activity"], target_lang="fr"), axis=1)
#activity_Datareport_French['Districts'] = activity_Datareport_French.apply(lambda row: model.translate(row["Districts"], target_lang="fr"), axis=1)

## Column rename
subactivity_Datareport_French = subactivity_Datareport_French.rename(columns={ 
                                                'Name of the Sub-activity': model.translate('Name of the Sub-activity', target_lang="fr"),
                                                'Total Program Costs (Sub-activity Costs)' : model.translate('Total Program Costs (Sub-activity Costs)', target_lang="fr"),
                                                'Amount of Funding Received' : model.translate('Amount of Funding Received', target_lang="fr") ,
                                                'Gap' : model.translate('Gap', target_lang="fr"),
                                                'Funding Received in Percentage' : model.translate('Funding Received in Percentage', target_lang="fr")})
subactivity_Datareport_French
```

```python
# Donor data in French
donor_Data

# Create french version
donor_Data_French = donor_Data.copy()
donor_Data_French['Donor'] = donor_Data_French.apply(lambda row: model.translate(row["Donor"], target_lang="fr"), axis=1)
#activity_Datareport_French['Districts'] = activity_Datareport_French.apply(lambda row: model.translate(row["Districts"], target_lang="fr"), axis=1)

## Column rename
donor_Data_French = donor_Data_French.rename(columns={ 
                                                'Donor': model.translate('Donor', target_lang="fr"),
                                                'Total cost of all programs in GNF' : model.translate('Total cost of all programs in GNF', target_lang="fr"),
                                                'Funding received in GNF' : model.translate('Funding received in GNF', target_lang="fr") ,
                                                'Funding gap in GNF' : model.translate('Funding gap in GNF', target_lang="fr"),
                                                'Funding Received in Percentage' : model.translate('Funding Received in Percentage', target_lang="fr")})
donor_Data
```

```python
# activity_Donationsbydonor
activity_Donationsbydonor_French = activity_Donationsbydonor.copy()

activity_Donationsbydonor_French['Activity'] = activity_Donationsbydonor_French.apply(lambda row: model.translate(row["Activity"], target_lang="fr"), axis=1)
#activity_Datareport_French['Districts'] = activity_Datareport_French.apply(lambda row: model.translate(row["Districts"], target_lang="fr"), axis=1)

## Column rename
activity_Donationsbydonor_French = activity_Donationsbydonor_French.rename(columns={ 
                                                'National Government': model.translate('National Government', target_lang="fr"),
                                                'USAID' : model.translate('USAID', target_lang="fr"),
                                                'Sightsavers' : model.translate('Sightsavers', target_lang="fr") ,
                                                'Fondation Raoul Follereau France' : model.translate('Fondation Raoul Follereau France', target_lang="fr"),
                                                'Speak Up Africa' : model.translate('Speak Up Africa', target_lang="fr"),
                                                'Ordre de Malte' : model.translate('Ordre de Malte', target_lang="fr"),
                                                'WHO' : model.translate('WHO', target_lang="fr"),
                                                'Catholic Relief Services' : model.translate('Catholic Relief Services', target_lang="fr"),
                                                'Institut de Recherche et de Développement France' : model.translate('Institut de Recherche et de Développement France', target_lang="fr"),
                                                'Institut Pasteur de Guinée' : model.translate('Institut Pasteur de Guinée', target_lang="fr")})
activity_Donationsbydonor_French
```

```python
# personnel_costsData in French
personnel_costsData_French = personnel_costsData

personnel_costsData_French['Personnel title'] = personnel_costsData_French.apply(lambda row: model.translate(row["Personnel title"], target_lang="fr"), axis=1)
personnel_costsData_French['Government employee\n(yes/no)'] = personnel_costsData_French.apply(lambda row: model.translate(row["Government employee\n(yes/no)"], target_lang="fr"), axis=1)
personnel_costsData_French['Full or part-time \nNTD staff'] = personnel_costsData_French.apply(lambda row: model.translate(row["Full or part-time \nNTD staff"], target_lang="fr"), axis=1)
personnel_costsData_French['Resource Location'] = personnel_costsData_French.apply(lambda row: model.translate(row["Resource Location"], target_lang="fr"), axis=1)

## Column rename
personnel_costsData_French = personnel_costsData_French.rename(columns={ 
                                                'If tracking salary,\nAnnual Salary': model.translate('If tracking salary,\nAnnual Salary', target_lang="fr"),
                                                'If full-time, how many \npeople under this title?' : model.translate('If full-time, how many people under this title?', target_lang="fr"),
                                                'Full or part-time \nNTD staff': model.translate('Full or part-time \nNTD staff', target_lang="fr"),
                                                'Personnel title': model.translate('Personnel title', target_lang="fr"),
                                                'Government employee\n(yes/no)': model.translate('Government employee\n(yes/no)', target_lang="fr"),
                                                'Resource Location': model.translate('Resource Location', target_lang="fr")})
personnel_costsData_French
```
