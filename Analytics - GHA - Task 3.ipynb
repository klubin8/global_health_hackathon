### Program costs, funding sources, and funding gaps disaggregated by activity, sub-activity and donor


* Activity related analysis

```python
## Activity related analysis
activity_Data = data[['Name of Activity.2','Amount of Finance Recevied','Gap']]

activity_Data = activity_Data.groupby(['Name of Activity.2'])['Amount of Finance Recevied','Gap'].sum().reset_index()

activity_Data['Total Program Costs (Activity Costs)'] = activity_Data[['Amount of Finance Recevied','Gap']].sum(axis=1)

activity_Datareport = activity_Data[['Name of Activity.2','Total Program Costs (Activity Costs)','Amount of Finance Recevied','Gap']]

activity_Datareport.rename(columns={'Name of Activity.2': 'Name of the Activity','Amount of Finance Recevied':'Amount of Funding Received'}, inplace=True)

activity_Datareport['Funding Received in Percentage'] = activity_Datareport['Amount of Funding Received']*100/activity_Datareport['Total Program Costs (Activity Costs)']

activity_Datareport
```

- Sub-activity related analysis

```python
#Select columns needed for sub-activity related analysis
subactivity_Data = data[['Name of Subactivity.1','Cost of Subactivity in GNF.1','Amount of Finance Recevied','Gap']]

#Select columns needed for sub-activity related analysis
subactivity_Data = data[['Name of Subactivity.1','Amount of Finance Recevied','Gap']]

subactivity_Data = subactivity_Data.groupby(['Name of Subactivity.1'])['Amount of Finance Recevied','Gap'].sum().reset_index()

subactivity_Data['Total Program Costs (Sub-activity Costs)'] = subactivity_Data[['Amount of Finance Recevied','Gap']].sum(axis=1)

subactivity_Datareport = subactivity_Data[['Name of Subactivity.1','Total Program Costs (Sub-activity Costs)','Amount of Finance Recevied','Gap']]

subactivity_Datareport.rename(columns={'Name of Subactivity.1': 'Name of the Sub-activity','Amount of Finance Recevied':'Amount of Funding Received'}, inplace=True)

subactivity_Datareport['Funding Received in Percentage'] = subactivity_Datareport['Amount of Funding Received']*100/subactivity_Datareport['Total Program Costs (Sub-activity Costs)']

subactivity_Datareport
```

- Donor Data Analysis

```python
donor_Data = data[[  'Cost of Subactivity in GNF.1', 'National Government', 'USAID', 'Sightsavers',
                     'Fondation Raoul Follereau France', 'Speak Up Africa', 'Ordre de Malte', 'WHO',
                     'Catholic Relief Services ','Institut de Recherche  et de Développement France ',
                     'Institut Pasteur  de  Guinée']]

donor_Data = donor_Data.sum().to_frame(name = None)#.rest_index()

donor_Data = donor_Data.rename(columns={0: 'Funding received in GNF'})

donor_Data[['Total cost of all programs in GNF']] = donor_Data[['Funding received in GNF']].iloc[0]

donor_Data['Funding gap in GNF'] = donor_Data['Total cost of all programs in GNF'] - donor_Data['Funding received in GNF']

donor_Data = donor_Data.iloc[1:,:]

#Re-arrange columns
donor_Data = donor_Data[['Total cost of all programs in GNF','Funding received in GNF','Funding gap in GNF']]

donor_Data = donor_Data.reset_index()

donor_Data = donor_Data.rename(columns={'index': 'Donor'})

donor_Data['Funding Received in Percentage'] = donor_Data['Funding received in GNF']*100/donor_Data['Total cost of all programs in GNF']

donor_Data
```
